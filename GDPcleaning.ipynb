```python
import pandas as pd
# Clone in git repository
! git clone https://github.com/annab0503/DS4002

# Load in the data
excel_data = pd.ExcelFile('/content/DS4002/Project 1/Original Data/WITS-Country-Timeseries.xlsx')


# Load a specific sheet (e.g., the first sheet) into a DataFrame
df = excel_data.parse('Country-Timeseries')  # Change index for other sheets

# Display the first few rows of the DataFrame
df.head()
```

```python
df_melted = df.melt(id_vars=["Country Name", "Indicator Name"],
                    var_name="Year",
                    value_name="GDP")

df_melted = df_melted.drop(columns=["Indicator Name"])
df_melted['Year'] = pd.to_numeric(df_melted['Year'], errors='coerce')

df_melted = df_melted[df_melted['Year'].between(1997, 2014)]

df_melted = df_melted.dropna(subset=["GDP"])

df_melted
```

```python

df_melted.to_csv('/content/cleaned_gdp_data.csv', index=False)
```
